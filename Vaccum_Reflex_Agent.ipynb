```python
# Import statements
import random
```

```python
# Defines the environment for the vacuum cleaning problem
class VacuumEnvironment():
    def __init__(self, status):
        self.status = status
        
    # returns what the agent can percieve in the current environment
    def percept(self, agent):
        return agent.location, self.status[agent.location]
    
    # Updates the location and status of the agent after it chooses an action    
    def action(self, agent, action):
        if action == "Right":
            agent.location = "Loc2"
            agent.performance -= 1
        if action == "Left":
            agent.location = "Loc1"
            agent.performance -= 1
        elif action == "Suck":
            if self.status[agent.location] == "Dirty":
                agent.performance -= 1
                self.status[agent.location] = "Clean"
            else:
                agent.performance -= 5
                
    # Defines the starting location of the agent
    def start_location(self):
        return random.choice(["Loc1", "Loc2"])
```

```python
# Defines a simple reflex agent to perform in the vacuum environment
class VacuumReflexAgent():
    def __init__(self, location, performance):
        self.location = location
        self.performance = performance
        
    # Chooses an action for the agent to perform based on its current perception of the environment
    def action_plan(self, percept_location, percept_status):
        if percept_status == "Dirty":
            action = "Suck"
        elif percept_location == "Loc1":
            action = "Right"
        else:
            action = "Left"
            
        return action
```

```python
# Defines the actions for an agent to execute
def execute_agent(configuration):
    
    if configuration == 0:
        initial_status = {
            "Loc1": "Clean",
            "Loc2": "Clean"
        }
        initial_location = "Loc1"
    elif configuration == 1:
        initial_status = {
            "Loc1": "Dirty",
            "Loc2": "Clean"
        }
        initial_location = "Loc1"
    elif configuration == 2:
        initial_status = {
            "Loc1": "Clean",
            "Loc2": "Dirty"
        }
        initial_location = "Loc1"
    elif configuration == 3:
        initial_status = {
            "Loc1": "Dirty",
            "Loc2": "Dirty"
        }
        initial_location = "Loc1"
    elif configuration == 4:
        initial_status = {
            "Loc1": "Clean",
            "Loc2": "Clean"
        }
        initial_location = "Loc2"
    elif configuration == 5:
        initial_status = {
            "Loc1": "Dirty",
            "Loc2": "Clean"
        }
        initial_location = "Loc2"
    elif configuration == 6:
        initial_status = {
            "Loc1": "Clean",
            "Loc2": "Dirty"
        }
        initial_location = "Loc2"
    else:
        initial_status = {
            "Loc1": "Dirty",
            "Loc2": "Dirty"
        }
        initial_location = "Loc2"
    
    env = VacuumEnvironment(initial_status)
    agent = VacuumReflexAgent(initial_location, 0)
   
    print(f"Configuration {configuration + 1} initial status: {env.status}")
    print(f"Configuration {configuration + 1} initial vacuum start: {agent.location}")
    while env.status["Loc1"] != "Clean" or env.status["Loc2"] != "Clean":
        location, status = env.percept(agent)
        action_plan = agent.action_plan(location, status)
        action = env.action(agent, action_plan)
        
    agent.performance += 10
    
    return agent.performance 
```

```python
config = 0
average_performance = 0
while config < 8:
    performance = execute_agent(config)
    print(f"Configuration {config+1} had performance score: {performance}")
    average_performance += performance
    config += 1
    
average_performance /= 8
print(f"Average performance score: {average_performance}")
```
